```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np
```

```python
MINX = -5
MAXX = - MINX
dt = 0.1
x = np.arange(MINX,MAXX,dt)
alpha = 0.05
N01 = norm.pdf(x,0,1) #H_0
rejection_point = norm.ppf(1-alpha/2)
rejectreg_right = np.arange(rejection_point, MAXX, dt)
rejectreg_left = - rejectreg_right
```

```python
plt.plot(x, N01, label="$Model: N(0,1)$", c='black')
plt.fill_between(rejectreg_right, norm.pdf(rejectreg_right), color='red', alpha=0.3, label="rejection region")
plt.fill_between(rejectreg_left, norm.pdf(rejectreg_left), color='red', alpha=0.3)
plt.scatter([-2.3],[0], c='red', label="$X$: reject")
plt.legend()
plt.xticks([-1.96, 1.96])
plt.yticks([])
plt.savefig('anomaly-detection.png')
```
