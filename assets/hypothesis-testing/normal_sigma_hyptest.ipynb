```python
#nbi:hide_in
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np
from ipywidgets import interact
```

```python
MAXX = 3
dt = 0.1
x = np.arange(-MAXX,MAXX,dt)
alpha = 0.05
```

```python
N01 = norm.pdf(x,0,1)
N02 = norm.pdf(x,0,0.25)
rejection_point = norm.ppf(1-alpha/2)
rejectreg_right = np.arange(rejection_point, MAXX, dt)
rejectreg_left = - rejectreg_right
```

```python
plt.plot(x, N01, label="$H_0: N(0,1)$")
plt.plot(x, N02, label="$H_1: N(0,0.25)$")
plt.fill_between(rejectreg_right, norm.pdf(rejectreg_right), color='blue', alpha=0.3, label="reject region")
plt.fill_between(rejectreg_left, norm.pdf(rejectreg_left), color='blue', alpha=0.3)
plt.legend()
```

```python
lik_ratio = N02/N01
plt.plot(x, lik_ratio, label="likelihood ratio")
plt.legend()
```
